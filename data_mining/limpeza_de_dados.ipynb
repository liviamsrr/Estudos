```python
import pandas as pd

# ==============================================================================
# 2. FUNÇÕES DE APOIO E DADOS DE PERCENTIS
# (Mesma lógica da resposta anterior)
# ==============================================================================

# --- Dados de Amostra (Extrato da Tabela de Percentis de IMC por Idade do CDC) ---
percentis_meninos = {
    24: {'p5': 14.7, 'p85': 17.3, 'p95': 18.1},  # 2 anos
    60: {'p5': 13.8, 'p85': 16.6, 'p95': 17.4},  # 5 anos
    120: {'p5': 14.2, 'p85': 18.6, 'p95': 20.4}, # 10 anos
    180: {'p5': 16.1, 'p85': 23.2, 'p95': 25.9}, # 15 anos
    228: {'p5': 17.8, 'p85': 26.2, 'p95': 29.4}, # 19 anos
}
percentis_meninas = {
    24: {'p5': 14.3, 'p85': 17.1, 'p95': 18.0},  # 2 anos
    60: {'p5': 13.6, 'p85': 16.7, 'p95': 17.7},  # 5 anos
    120: {'p5': 14.0, 'p85': 18.9, 'p95': 21.1}, # 10 anos
    180: {'p5': 16.3, 'p85': 24.1, 'p95': 27.2}, # 15 anos
    228: {'p5': 17.6, 'p85': 26.5, 'p95': 30.1}, # 19 anos
}

def classificar_imc_crianca_para_df(linha):
    """
    Função projetada para ser usada com df.apply().
    Ela recebe uma linha do DataFrame como entrada.
    """
    # Extrai os dados da linha
    sexo = linha['sexo']
    idade_anos = linha['idade_anos']
    imc = linha['imc']
    
    # Validações básicas
    if pd.isna(sexo) or pd.isna(idade_anos) or pd.isna(imc) or imc <= 0:
        return "Dados insuficientes"

    # Converte idade para meses totais
    idade_total_meses = idade_anos * 12

    # Seleciona a tabela de percentis correta
    tabela_percentis = percentis_meninos if sexo.lower() == 'masculino' else percentis_meninas

    # Encontra a idade mais próxima na nossa amostra de dados
    idade_chave_mais_proxima = min(tabela_percentis.keys(), key=lambda k: abs(k - idade_total_meses))
    
    percentis = tabela_percentis[idade_chave_mais_proxima]
    p5, p85, p95 = percentis['p5'], percentis['p85'], percentis['p95']
    
    # Realiza a classificação
    if imc < p5:
        return "Abaixo do peso"
    elif p5 <= imc < p85:
        return "Peso normal"
    elif p85 <= imc < p95:
        return "Sobrepeso"
    else: # imc >= p95
        return "Obesidade"

# ==============================================================================
# 3. CALCULAR A COLUNA 'imc' NO DATAFRAME
# ==============================================================================
# Esta operação é "vetorizada", ou seja, muito rápida e eficiente no pandas.
df['imc'] = df['peso_kg'] / (df['altura_m'] ** 2)

# ==============================================================================
# 4. APLICAR A FUNÇÃO DE CLASSIFICAÇÃO E CRIAR A NOVA COLUNA
# ==============================================================================
# O método .apply(funcao, axis=1) executa a função 'funcao' em cada linha do DataFrame.
# 'axis=1' especifica que a operação deve ser feita por linha.
df['classificacao_imc'] = df.apply(classificar_imc_crianca_para_df, axis=1)

# ==============================================================================
# 5. EXIBIR O RESULTADO
# ==============================================================================
print("DataFrame com IMC calculado e classificado:")
# Arredondando o IMC para melhor visualização
print(df.round({'imc': 2}))
```
